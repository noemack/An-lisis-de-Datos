# Análisis de Datos - Turismo Uruguay - Cruceros

Conjunto referente a los datos obtenidos de encuestas de cruceros, determinando las principales características de sus viajes, así como también estimar el gasto que realizan durante el mismo. Los datos son referentes a cada temporada a partir del año 2016.

La información que se encuentra detallada es sobre; fecha del viaje, nacionalidad del crucerista, puerto de arribo, cantidad de personas en el grupo de viaje, y los distintos gastos realizados entre otros.

https://catalogodatos.gub.uy/dataset/ministerio-de-turismo-turismo-cruceros

El recurso cuenta con los datos desagregados obtenidos de Cruceros, determinando las principales características de sus viajes, así como también estimar el gasto que realizan durante el mismo.

Atributos:
* IdCruceros: Es el código identificatorio de los Cruceros
* idNacionalidad: Es el código de la nacionalidad del entrevistado
* Pais: Es la Nacionalidad del entrevistado
* Fecha: Fecha realizada la encuesta
* idfecha: Es el código de la fecha de realizada la encuesta
* idPuerto: Es el código identificatorio del puerto de arribo del crucerista
* Puerto: Es el Puerto de arribo del crucerista
* TotalPersonas: Cantidad de Personas en cada grupo
* CantidadVisitas: Cantidad de vistias anteriores
* GastoTotal: Gasto para total, el valor de este gasto es medido en Dólares Americanos
* GastoTours: Gasto para Tours, el valor de este gasto es medido en Dólares Americanos
* GastoAlimentacion: Gasto para la Alimentación, el valor de este gasto es medido en Dólares Americanos
* GastoTransporte: Gasto para el Transporte, el valor de este gasto es medido en Dólares Americanos
* GastoShopping: Gasto para Shopping, el valor de este gasto es medido en Dólares Americanos
* GastoOtros: Gasto para Otros, el valor de este gasto es medido en Dólares Americanos

Gastos: se puede detectar que existen varias personas que no tienen registros de gastos en transporte, alimentación, ni otros.

```python
# filtremos por GastoShopping, para saber cuantos regitros son mayores a 0
cruceros_filtrado = cruceros[cruceros["GastoShopping"] >0]
cruceros_filtrado
```

```python
# filtremos los registros que tienen GastoTransporte, GastoAlimentacion, GastoOtros en 0
cruceros_filtrado = cruceros[(cruceros["GastoTransporte"] == 0)  & (cruceros["GastoAlimentacion"] == 0) & (cruceros["GastoOtros"] == 0)]
cruceros_filtrado
```

```python
# otra forma de filtrar las variables en 0

# Lista de variables que deben ser igual a cero
variables_cero = ["GastoTransporte", "GastoAlimentacion", "GastoOtros"]

# Condiciones para cada variable en la lista
condiciones = [cruceros[var] == 0 for var in variables_cero]

# Combinar las condiciones usando la función reduce
from functools import reduce

# Aplicar todas las condiciones usando reduce y el operador lógico AND
condicion_general = reduce(lambda x, y: x & y, condiciones)

# Filtrar el DataFrame
cruceros_filtrado = cruceros[condicion_general]
cruceros_filtrado
```

## 3- Exploración de los datos y visualizaciones

```python
# Exploración
total_cruceros_por_pais = cruceros.groupby('Pais')['IdCruceros'].count()
```

```python
total_cruceros_por_pais
```

mmm... qué es ese país que se llama "Sin Datos"?

```python
# Visualización
plt.figure(figsize=(10, 6))
total_cruceros_por_pais.plot(kind='bar')
plt.title('Total de cruceros por Pais')
plt.xlabel('Pais')
plt.ylabel('Total de Cruceros')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
